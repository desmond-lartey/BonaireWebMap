```python
import osmnx as ox

# Geocode to get the boundary polygon for Budapest
admin = ox.geocode_to_gdf('Bonaire')
admin_poly = admin.geometry.values[0]

# Create a graph from the polygon
G = ox.graph_from_polygon(admin_poly)

# Get the number of nodes and edges
G.number_of_nodes(), G.number_of_edges()

# Convert the graph to GeoDataFrames
nodes, edges = ox.graph_to_gdfs(G)

# Plot the graph with a dark background
import matplotlib.pyplot as plt
plt.style.use('dark_background')

# Create the plot
f, ax = plt.subplots(1, 1, figsize=(10, 10))
edges.plot(ax=ax, edgecolor='white', linewidth=0.3)
ax.axis('off')
```

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Geocode to get the boundary polygon for Budapest
admin = ox.geocode_to_gdf('Budapest')
admin_poly = admin.geometry.values[0]

# Create a graph from the polygon
G = ox.graph_from_polygon(admin_poly)

# Convert the graph to GeoDataFrames
nodes, edges = ox.graph_to_gdfs(G)

# Plot the graph with a dark background
plt.style.use('dark_background')

# Create the plot
f, ax = plt.subplots(1, 1, figsize=(10, 15))
edges.plot(column='length', ax=ax, cmap='Reds', alpha=0.7)
nodes.plot(ax=ax, cmap='Blues', alpha=0.6)

# Add a basemap
ctx.add_basemap(ax, alpha=0.96, crs=4326, url=ctx.providers.Stamen.TonerLite)
ax.axis('off')
```
